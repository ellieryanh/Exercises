```python
import tensorflow.keras.preprocessing.image as img_to_array
import matplotlib.pyplot as plt
import numpy as np
import os
from os import listdir
from google.colab.patches import cv2_imshow
import cv2
```

```python
#import tensorflow as tf
import tensorflow.keras.preprocessing.image as img_to_array
# get the path/directory
path = "/content/TRY "

#MAKE FUNCTION FOR THIS TO SHOW IMAGES

data = []
label=[]
for images in os.listdir(path):
 
    # check if the image ends with png
    if (images.endswith(".png")):
        print(images)
        path_new='/content/TRY /'+images
        img=cv2.imread(path_new)
        cv2_imshow(img)
        print('Original Dimensions : ',img.shape)
        dim = (28, 28)
        resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
        print('Resized Dimensions : ',resized.shape)
        cv2_imshow(resized)
       # img_array = img_to_array(resized)
        label.append(0)
        data.append(resized)






#MAKE THIS FUNCTION RESIZING AND ADDING TO ARRAY
```
